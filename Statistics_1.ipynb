Question 1: What is a random variable in probability theory?

Answer:
A random variable is a variable whose value is determined by the outcome of a random experiment. It assigns numerical values to different outcomes.
For example, if you toss a coin, you can define a random variable X such that Heads = 1 and Tails = 0.

Question 2: What are the types of random variables?

Answer:
There are two main types:

Discrete random variable: Takes countable values such as 0,1,2,...

Continuous random variable: Takes infinitely many values within an interval, such as height, weight, time.

Question 3: Explain the difference between discrete and continuous distributions.

Answer:
A discrete distribution deals with outcomes that can be counted, like number of customers in a shop.
A continuous distribution deals with outcomes that can take any real value in an interval, like temperature or rainfall.
Discrete values are exact, continuous values are measured.

Question 4: What is a binomial distribution, and how is it used in probability?

Answer:
A binomial distribution represents the number of successes in a fixed number of independent trials where each trial has two outcomes (success or failure).
It is used when:
• number of trials is fixed
• probability of success remains constant
• trials are independent
Example: number of heads in 10 coin tosses.

Question 5: What is the standard normal distribution, and why is it important?

Answer:
The standard normal distribution is a normal distribution with mean 0 and standard deviation 1.
It is important because many statistical tests and probability calculations use it as a reference model. It allows us to convert any normal variable into a Z-score for easy comparison.

Question 6: What is the Central Limit Theorem (CLT), and why is it critical in statistics?

Answer:
The Central Limit Theorem states that when we take repeated samples from any population, the distribution of the sample mean becomes approximately normal as the sample size increases, even if the original data are not normally distributed.
It is important because it allows us to use normal distribution methods for confidence intervals and hypothesis tests.

Question 7: What is the significance of confidence intervals in statistical analysis?

Answer:
A confidence interval gives a range of values that is likely to contain the true population mean.
It shows the level of uncertainty in estimation.
For example, a 95% confidence interval means we are 95% confident that the true mean lies within the calculated range.

Question 8: What is the concept of expected value in a probability distribution?

Answer:
The expected value is the long-run average value of a random variable.
It is calculated by multiplying each outcome with its probability and adding them.
It represents the mean of the probability distribution.

Question 10: Applying Central Limit Theorem to sales data
Explanation

We have daily sales data. To estimate the average sales for the population:

Compute the sample mean.

Compute the sample standard deviation.

Use the Central Limit Theorem to calculate a 95% confidence interval:

Confidence Interval = mean ± 1.96 × (std / √n)

This gives a range where the true average daily sales likely fall.

```python
import numpy as np

daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

data = np.array(daily_sales)

mean_sales = np.mean(data)
std_sales = np.std(data, ddof=1)  # sample standard deviation
n = len(data)

# 95% confidence interval
margin_of_error = 1.96 * (std_sales / np.sqrt(n))
lower_bound = mean_sales - margin_of_error
upper_bound = mean_sales + margin_of_error

print("Mean Sales:", mean_sales)
print("95% Confidence Interval:", (lower_bound, upper_bound))
Sample Output:
Mean Sales: 248.7
95% Confidence Interval: (242.36, 255.04)

If you want, I can format this entire assignment as a clean PDF or Word file.







```
